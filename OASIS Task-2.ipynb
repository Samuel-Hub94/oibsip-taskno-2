```python
import pandas as pd
import plotly.express as px

# Load the data from the CSV file
df = pd.read_csv('Summary( Unemployment in India).csv')

# Define the order of the months
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# Convert the 'Month' column to categorical data type with the custom sort order
df['Month'] = pd.Categorical(df['Month'], categories=Month_order, ordered=True)

# Calculate the mean Estimated Unemployment Rate for each month and location
monthly_data = df.groupby(['Month', 'Location'])['Estimated Unemployment Rate (%)'].mean().reset_index()

# Create line plots for each location
fig = px.line(monthly_data, x='Month', y='Estimated Unemployment Rate (%)', color='Location',
              title='Estimated Unemployment Rate for Each Location (2019-2020)',
              labels={'Estimated Unemployment Rate (%)': 'Unemployment Rate (%)'})

# Adjust the layout for better visibility
fig.update_layout(xaxis={'title': 'Month'}, yaxis={'title': 'Unemployment Rate (%)'})

# Show the plot
fig.show()
```

<h3>Unemployment rate before and after Lockdown</h3>

```python
# data representation before and after lockdown

before_lockdown = df[(df['month_int']>=1) &(df['month_int'] <4)]
after_lockdown = df[(df['month_int'] >=4) & (df['month_int'] <=6)]
af_lockdown = after_lockdown.groupby('Location')['Estimated Unemployment Rate (%)'].mean().reset_index()

lockdown = before_lockdown.groupby('Location')['Estimated Unemployment Rate (%)'].mean().reset_index()
lockdown['unemployment rate before lockdown'] = af_lockdown['Estimated Unemployment Rate (%)']

lockdown.columns = ['Location','unemployment rate before lockdown','unemployment rate after lockdown']
lockdown.head()
```

```python
# unenployment rate change after lockdown

lockdown['rate change in unemployment'] =round(lockdown['unemployment rate before lockdown']-lockdown['unemployment rate before lockdown']
                                              /lockdown['unemployment rate after lockdown'],2)
fig = px.bar(lockdown,x='Location',y='rate change in unemployment',color='rate change in unemployment',
            title='Percentage change in Unemployment rate in each state after lockdown',template='ggplot2')
fig.update_layout(xaxis={'categoryorder':'total ascending'})
fig.update_layout(width=800, height=600)
fig.show()
```

```python
import pandas as pd
import plotly.express as px

# Load your dataset into a DataFrame (replace 'your_data.csv' with your actual file)
df = pd.read_csv('Summary( Unemployment in India).csv')

# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Create a 'Year' column
df['Year'] = df['Date'].dt.year

# Extract month from 'Date' column
df['Month'] = df['Date'].dt.month

# Filter data for the years 2019 and 2020
df_2019_2020 = df[df['Year'].isin([2019, 2020])]

# Group by 'Month' and 'Year' and find the maximum unemployment rate for each month
max_unemployment_by_month = df_2019_2020.groupby(['Month', 'Year'])['Estimated Unemployment Rate (%)'].max().reset_index()

# Create a bar chart for each year
fig = px.bar(max_unemployment_by_month, x='Month', y='Estimated Unemployment Rate (%)',
             color='Year', facet_col='Year', facet_col_wrap=1,
             labels={'Estimated Unemployment Rate (%)': 'Unemployment Rate (%)'},
             title='Highest Unemployment Rate by Month for 2019 and 2020')

fig.update_xaxes(title='Month', tickmode='array', tickvals=list(range(1, 13)),
                 ticktext=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
fig.update_yaxes(title='Unemployment Rate (%)')

fig.show()
```
